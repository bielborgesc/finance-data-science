### Database with more actions

```python
#BOVA11: https://www.sunoresearch.com.br/artigos/bova11/
def getStock(stockName, dateStart = '2015-01-01'):
    ticker = yf.Ticker(stockName)
    stock_df = ticker.history(start=dateStart)
    return stock_df 
    
stocks_df = pd.DataFrame()
stocks = ['GOL', 'APLE', 'PLTR', 'KO', 'VZ', 'AMC']

for stock in stocks:
    stocks_df[stock] = getStock(stock)['Close']
```

```python
stocks_df #show database
```

```python
#Verify null values
stocks_df.isnull().sum()
```

```python
stocks_df.shape
```

```python
#Remove null values
stocks_df.dropna(inplace=True)
```

```python
stocks_df.shape
```

```python
stocks_df.isnull().sum()
```

```python
stocks_df.to_csv('stocks.csv')
```
